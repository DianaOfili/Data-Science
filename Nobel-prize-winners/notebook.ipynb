The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

```python
us_df = df.copy()
us_df['us_winner'] = us_df['birth_country'] == 'United States of America'
us_df['decade'] = (np.floor(us_df['year'] / 10) * 10).astype('int')
```

```python
us_df_prop = us_df.groupby('decade', as_index=False)['us_winner'].mean()
```

```python
max_us = us_df_prop['us_winner'].max()
max_decade_usa = us_df_prop[us_df_prop['us_winner'] == max_us]['decade'].values[0]
max_decade_usa
```

```python
female_df = df.copy()
female_df['decade'] = (np.floor(female_df['year'] / 10) * 10).astype('int')
female_df['female_laureates'] = female_df['sex'] == 'Female'
```

```python
female_df_prop = female_df.groupby(['decade', 'category'], as_index=False)['female_laureates'].mean()
female_max_prop = female_df_prop['female_laureates'].max()
female_decade_cat_max = female_df_prop[female_df_prop['female_laureates'] == female_max_prop][['decade', 'category']]
```

```python
max_female_dict = {female_decade_cat_max['decade'].values[0]:female_decade_cat_max['category'].values[0]}
```
